```python
import numpy as np
import pandas as pd
import geopandas as gpd
import pysal as ps
```

```python
def dbf2DF(dbfile, geographic_boundary, year): #Reads in DBF files and returns Pandas DF
        # you must put 'year' and 'geographic_boundary' in quotes
        # geographic_boundary is the state you have data for or "AllStates"
    db = ps.open(dbfile) #Pysal to open DBF
    d = {col: db.by_col(col) for col in db.header} #Convert dbf to dictionary
    pandasDF = pd.DataFrame(d) #Convert to Pandas DF
    db.close()
    pandasDF.to_csv('Pres_bycounty_' + geographic_boundary + '_' + year + '.csv')
    return pandasDF
```

```python
dbf2DF('elpo04p020.dbf', 'AllStates', '2004')
```
